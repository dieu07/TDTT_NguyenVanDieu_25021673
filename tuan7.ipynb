```python
#B1: Viết hàm tìm kiếm nhị phân trên một danh sách đã
#được sắp xếp tăng dần.Trả về chỉ số của phần tử
#hoặc -1 nếu không tìm thấy. mẫu hàm: def binary_search(arr, left, right, target):
def binary_search(arr,target):
    left, right = 0, len(arr) -1
    while left <= right:
        mid = (left + right)//2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1
arr = list(map(int,input().split()))
target = int(input())
print(binary_search(arr,target))
```

```python
#B2: Cho danh sách arr và một số nguyên x, đếm xem x xuất hiện bao nhiêu lần
#trong danh sách. def count_occurrences(arr, x) -> int:
def count_occurren(arr,x):
    count = 0
    for i in arr:
        if i == x:
            count += 1
    return count
arr = list(map(int,input().split()))
x = int(input())
print(count_occurren(arr,x))
```

```python
#B3: Viết hàm thực hiện thuật toán bubble sort, Selection Sort để sắp xếp danh sách
#theo thứ tự tăng dần. Sắp xếp nổi bọt (Bubble Sort); 
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        swapped = False
        for j in range(0,n-i-1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                swapped = True
        if not swapped:
            break
    return arr
arr = list(map(int,input().split()))
print(bubble_sort(arr))
```

```python
#B3: Selection-Sort
def selection_sort(arr):
    n = len(arr)
    for i in range(n):
        min_index = i
        for j in range(i+1,n):
            if arr[j] < arr[min_index]:
                min_index = j
        arr[i], arr[min_index] = arr[min_index], arr[i]
    return arr
arr = list(map(int,input().split()))
print(" ".join(map(str,selection_sort(arr))))
```

```python
#B4: Viết hàm tìm phần tử có số lần xuất hiện nhiều nhất trong danh sách. Nếu có
#nhiều phần tử cùng số lần,trả về phần tử xuất hiện trước.
def count_max(arr):
    d = {}
    for i in arr:
        if i not in d:
            d[i] = 1
        else:
            d[i] += 1
    displaymax = max(d.values())
    for key in d.keys():
        if d[key] == displaymax:
            print(f"{key} xuat hien nhieu nhat, som nhat, {displaymax} lan")
            break
arr = list(map(int,input().split()))
count_max(arr)
```

```python
#B6: Bạn được cung cấp một danh sách các danh sách lồng nhau
#  vd [[1, 2, 3], [4,5], [6, 7, 8, 9]], hãy làm phẳng danh sách
def lamphang(arr):
    s = []
    for i in range(len(arr)):
        if i%2 == 0:
            for j in arr[i]:
                s.append(j)
        else:
            for j in range(len(arr[i])-1,-1,-1):
                m = arr[i][j]
                s.append(m)
    return s
arr = eval(input())
print(lamphang(arr))
```

```python
#B7: Cho một danh sách các số nguyên nums, hãy tìm độ
# dài của dãy con tăng nghiêm ngặt dài nhất (không nhất thiết phải liền kề).
def lengthOfLIS(nums):
    if not nums:
        return 0

    n = len(nums)
    dp = [1] * n  # mỗi phần tử ít nhất tạo được dãy dài 1

    for i in range(n):
        for j in range(i):
            if nums[j] < nums[i]:
                dp[i] = max(dp[i], dp[j] + 1)

    return max(dp)
nums = list(map(int, input().split()))
print(lengthOfLIS(nums))
```

```python
#b8: Bạn được cung cấp: một danh sách các khoảng thời gian interval = [[l1, r1], [l2, r2], ...]
#  và một danh sách các truy vấn queries = [q1, q2, ...] Với mỗi truy vấn q, hãy tìm độ dài nhỏ nhất của
# khoảng thời gian [li, ri] sao cho li ≤ q ≤ ri. Nếu không tồn tại khoảng
# thời gian nào như vậy, hãy trả về -1 cho truy vấn đó. solve in < O(m.n), hint:
#binary search in the queries list
def interval(inter, que):
    res = []

    for q in que:
        ans = float('inf')
        found = False

        for l, r in inter:
            if l <= q <= r:
                found = True
                ans = min(ans, r - l + 1)

        if found:
            res.append(ans)
        else:
            res.append(-1)

    return res


inter = eval(input())
que = eval(input())
print(interval(inter, que))
```

```python
#Cách TỐI ƯU (< O(m·n))
import heapq

def interval(inter, que):
    inter.sort()
    qs = sorted((q, i) for i, q in enumerate(que))

    res = [-1] * len(que)
    heap = []
    i = 0

    for q, idx in qs:
        while i < len(inter) and inter[i][0] <= q:
            l, r = inter[i]
            heapq.heappush(heap, (r - l + 1, r))
            i += 1

        while heap and heap[0][1] < q:
            heapq.heappop(heap)

        if heap:
            res[idx] = heap[0][0]

    return res


inter = eval(input())
que = eval(input())
print(interval(inter, que))
```

```python
def behon(nums):
    s = []
    for i in range(len(nums)):
        count = 0
        for j in nums[i+1:]:
            if j < nums[i]:
                count += 1
        s.append(count)
    return s
nums = list(map(int,input().split()))
print(behon(nums))
```

```python
import bisect

def count_smaller(nums):
    sorted_list = []
    res = []

    for x in reversed(nums):
        idx = bisect.bisect_left(sorted_list, x)
        res.append(idx)
        bisect.insort(sorted_list, x)

    return res[::-1]

nums = list(map(int, input().split()))
print(count_smaller(nums))
```

Cho một danh sách các số
nguyên nums và một số
nguyên m, hãy tìm tổng
mảng con lớn nhất theo
modulo m.
Tức là, với bất kỳ mảng
con nums[i:j] nào, hãy tính
sum(nums[i:j]) % m và trả
về kết quả lớn nhất có thể.
Gợi ý:
Duy trì tổng tiền tố (prefix
sum) mod m và sử dụng
tìm kiếm nhị phân để tìm
tổng tiền tố nhỏ nhất chỉ
lớn hơn tổng hiện tại một
chút (để tối đa hóa hiệu số
bao quanh).

```python
import bisect

def max_subarray_mod(nums, m):
    prefix_list = []
    current = 0
    best = 0

    for x in nums:
        current = (current + x) % m
        best = max(best, current)

        # tìm prefix nhỏ nhất > current
        idx = bisect.bisect_right(prefix_list, current)
        if idx < len(prefix_list):
            best = max(best, (current - prefix_list[idx] + m) % m)

        bisect.insort(prefix_list, current)

    return best


nums = list(map(int, input().split()))
m = int(input())
print(max_subarray_mod(nums, m))
```

Dãy hài hòa là một dãy
thỏa mãn điều kiện giá trị
lớn nhất và giá trị nhỏ nhất
của dãy khác nhau đúng
bằng \( 1 \).
Viết hàm trả về độ dài của
dãy hài hòa con dài nhất có
thể có trong mảng giá trị
nguyên \( arr \) có \( n \)
phần tử
Ví dụ:
Input: 1 3 2 2 5 2 3 7
Output: 5 - Dãy hài hòa con
dài nhất là [3,2,2,2,3]


```python
from collections import Counter
def haihoa(arr):
    cnt = Counter(arr)
    best = 0
    for i in cnt:
        if i+1 in cnt:
            best = max(best, cnt[i] + cnt[i+1])
    return best
arr = list(map(int,input().split()))
print(haihoa(arr))
```
