```python
#b13
kwh = float(input("Nhập số kWh: "))
if kwh <= 50:
    tien = kwh * 1500
elif kwh <= 100:
    tien = 50 * 1500 + (kwh - 50) * 2000
else:  # kwh > 100
    tien = 50 * 1500 + 50 * 2000 + (kwh - 100) * 3000
print("Tiền điện:", int(tien), "đồng")
```
