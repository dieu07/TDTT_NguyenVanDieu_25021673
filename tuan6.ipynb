```python
#b4: Nhập một dòng gồm các cặp key:value cách nhau
# bởi dấu cách. Tạo dictionary ánh xạ mỗi key -> list
# các value theo thứ tự xuất hiện. Nếu key lặp lại,
# thêm vào list của key đó
n = input().split()
d = {}
for i in n:
    key, value = i.split(':')
    if key not in d:
        d[key] = [value]
    else:
        d[key].append(value)
print(d)
```

```python
#b5: Nhập một list số nguyên, In ra dictionary {'positives':..., 'negatives': ..., 'zeros': ...} (đếm số >0, <0, =0)
s = list(map(int,input().split()))
d = {'positives':[], 'negatives':[], 'zeros':[]}
for i in s:
    if i > 0:
        d['positives'].append(i)
    elif i < 0:
        d['negatives'].append(i)
    else:
        d['zeros'].append(i)
m = {'positives':len(d['positives']), 'negatives': len(d['negatives']), 'zeros':len(d['zeros'])}
print(m)
```

```python
#b10: Viết chương trình nhập vào một danh sách các số nguyên,
# và số nguyên dương k. In ra list gồm các cặp số khác nhau có tổng bằng k. Mỗi cặp là một tuple, hai số trong
# tuple có thể giống nhau, thứ tự xuất hiện trong danh sách của số thứ nhất nhỏ hơn thứ tự của số thứ hai. Các tuple
# trong list được sắp xếp theo thứ tự tăng dần của phần tử đầu tiên trong tuple.
nums = list(map(int, input().split()))
k = int(input())

seen = set()
result = []

for i in range(len(nums)):
    for j in range(i+1, len(nums)):
        if nums[i] + nums[j] == k:
            pair = (nums[i], nums[j])
            if pair not in seen:
                result.append(pair)
                seen.add(pair)

result.sort(key=lambda x: x[0])
print(result)
```

```python
#b11: Nhập vào một tuple chứa các số nguyên. Hãy tạo ra
#hai tuple mới: một chứa tất cả các số chẵn và một
#chứa tất cả các số lẻ từ tuple ban đầu, giữ nguyên
#thứ tự tương đối. In ra hai tuple này trên hai dòng
#riêng biệt
nums = tuple(map(int, input("Nhập các số nguyên, cách nhau bằng dấu cách: ").split()))
even_tuple = tuple(x for x in nums if x % 2 == 0)
odd_tuple = tuple(x for x in nums if x % 2 != 0)

print(even_tuple)
print(odd_tuple)
```

```python
#b12: [Tìm phần tử xuất hiện nhiều nhất]
#Nhập vào một list các số nguyên. Hãy tìm số xuất
#hiện nhiều lần nhất trong list. Nếu có nhiều số cùng
#có số lần xuất hiện nhiều nhất, hãy in ra số nhỏ nhất trong số đó.
nums = list(map(int,input().split()))
s = {}
for i in nums:
    if i not in s:
        s[i] = 1
    else:
        s[i] += 1
largest = 0
for i in s.values():
    if i > largest:
        largest = i
m = []
for x, y in s.items():
    if y == largest:
        m.append(x)
print(min(m))
```

```python
#b15: [Lọc dictionary theo giá trị] Nhập vào một dictionary có key là chuỗi và value là
#số nguyên, và một số nguyên k ở dòng tiếp theo.
#Hãy tạo một dictionary mới chỉ chứa các cặp keyvalue từ dictionary ban đầu mà có value lớn hơn k.
#In ra dictionary mới.
s = input().split()
k = int(input())
d = {}
for i in range(0,len(s)-1,2):
    d[s[i]] = int(s[i+1])
new = {}
for m, n in d.items():
    if n > k:
        new[m] = n
print(new)
```

```python
#b16: Nhậ̣p vào mảng n hàng, m cột số nguyên. In lại mảng ra màn hình như ma trận hai chiều, định dạng
#mỗi phần tử 4 vị trí căn phải

n, m = map(int, input("Nhập n và m: ").split())
matrix = []

for i in range(n):
    row = list(map(int, input().split()))
    matrix.append(row)

for row in matrix:
    for num in row:
        print(f"{num:>4}", end="")  # căn phải trong 4 ký tự
    print()
```

```python
#b17: Nhập vào mảng vuông 2 chiều n hàng, n cột, in ra đường chéo chính và đường chéo phụ của mảng
n = int(input("Nhập n: "))

matrix = []
for i in range(n):
    row = list(map(int, input().split()))
    matrix.append(row)

print("Đường chéo chính:", end=" ")
for i in range(n):
    print(matrix[i][i], end=" ")
print()

print("Đường chéo phụ:", end=" ")
for i in range(n):
    print(matrix[i][n-1-i], end=" ")
print()
```

```python
#b18: Nhậ̣p vào mảng n hàng, m cột số nguyên và số nguyên k (n, m, k nhập trên cùng một dòng, cách
#nhau dấu cách). Tính và in ra tổng các phần tử trên cột k (hàng k) 

n, m, k = map(int, input("Nhập n m k: ").split())
matrix = []
for i in range(n):
    row = list(map(int, input().split()))
    matrix.append(row)

# Tính tổng cột k
col_sum = sum(matrix[i][k] for i in range(n))

# Tính tổng hàng k
row_sum = sum(matrix[k])

print("Tổng cột", k, ":", col_sum)
print("Tổng hàng", k, ":", row_sum)
```

```python
#b19: Viết hàm đưa vào 1 list số nguyên, tìm và trả về vị trí có giá trị lớn nhất trong list
def vi_tri_max(lst):
    """
    Nhận vào một list số nguyên lst
    Trả về vị trí (index) của phần tử lớn nhất
    """
    if not lst:  # kiểm tra list rỗng
        return None

    max_val = lst[0]  # giả sử phần tử đầu là lớn nhất
    max_index = 0     # vị trí hiện tại

    for i in range(1, len(lst)):
        if lst[i] > max_val:
            max_val = lst[i]
            max_index = i

    return max_index
```

```python
#b20:
def tim_vi_tri(lst, k):
    """
    Nhận vào:
      - lst: list số nguyên
      - k: số nguyên dương cần tìm
    Trả về:
      - vị trí (index) phần tử đầu tiên có giá trị k
      - nếu không tìm thấy, trả về -1
    """
    for i in range(len(lst)):
        if lst[i] == k:
            return i  # tìm thấy → trả về vị trí
    return -1  # không tìm thấy
```
