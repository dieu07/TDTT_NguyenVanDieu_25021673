W8A3
[ValueError - HappyNumbers]
Nhập vào một số nguyên dương a, Kiểm tra xem
a có phải là số hạnh phúc không.
Một số nguyên dương n được gọi là số hạnh phúc
nếu ta thực hiện liên tiếp các bước sau:
- Tính tổng bình phương các chữ số của n.
- Lặp lại bước đó với kết quả vừa tính.
- Nếu cuối cùng kết quả là 1, thì n là số hạnh
phúc.
- Nếu rơi vào vòng lặp vô hạn không bao giờ ra 1
(ví dụ như 4 → 16 → 37 → ...), thì n không phải là
số hạnh phúc.
Hãy nhập thử đầu vào là "a" rồi đọc kỹ Traceback
để gỡ lỗi nhập liệu. (Dùng try ... except bắt lỗi
ValueError thêm set)

```python
n = input()
s = 0
for i in n:
     s += (int(i))**2
try:
    seen = set()

    if s == 1:
        print(s)
    else:
        while s != 1:
            if s in seen:
                raise ValueError("Vòng lặp vô hạn")
            else:
                seen.add(s)
            snew = 0
            scopy = str(s)            #for i in str(s):
            for i in scopy:
                snew += (int(i))**2
            s = snew
    print("YES")
except ValueError:
    print("NO")
```

W8A4
Sử dụng try except,kết hợp cùng else... finally.
Tính tổng 2 số, nếu nhập sai kiểu thì thông báo,
luôn luôn in "Kết thúc chương trình." sau khi chạy.

```python
try:
    a = float(input("Số thứ nhất: "))
    b = float(input("Số thứ hai: "))
    print("Tổng =", a + b)
except ValueError:
    print("Vui lòng nhập số hợp lệ!")
finally:
    print("Kết thúc chương trình.")
```

W8A5
Viết chương trình chuyển tất cả phần tử của một
list cho trước về int.
Nếu phần tử nào trong list không thể chuyển được
thì in cảnh báo và bỏ qua phần tử đó

```python
arr = input().split()

result = []

for x in arr:
    try:
        num = int(x)
        result.append(num)
    except ValueError:
        print(f"Canh bao: '{x}' khong the chuyen sang int, bo qua")

print("Danh sach sau khi chuyen:", result)
```

W10A6
Viết chương trình nhập một số x và tính căn bậc
hai của nó. Bắt lỗi ngoại lệ khi x là số âm. Kết quả
in ra chính xác tới 2 chữ số thập phân nếu x hợp
lệ. (Dùng try ... except bắt lỗi ValueError in ra "So
am") 

```python
import math

try:
    x = float(input())
    if x < 0:
        raise ValueError
    result = math.sqrt(x)
    print(f"{result:.2f}")
except ValueError:
    print("So am")
```

W10A10
Viết chương trình nhập vào phần tử N (N > 0) và
một mảng gồm N số nguyên dương. Một mảng là
hợp lệ nếu nó gồm các số khác nhau. In ra tổng
các phần tử nếu mảng đó là hợp lệ. Nếu mảng
không hợp lệ, in ra "Mang khong hop le".


```python
try:
    N = int(input())
    arr = list(map(int, input().split()))

    if N <= 0 or len(arr) != N:
        print("Mang khong hop le")
    elif len(set(arr)) != N:
        print("Mang khong hop le")
    else:
        print(sum(arr))
except:
    print("Mang khong hop le")
```

W8A11
Cho string S, kiểm tra xem S có phải một regex
hợp lệ hay không.
Đầu vào: Dòng đầu tiên chứa số nguyên T - số
lượng testcase. T dòng tiếp theo mỗi dòng chứa
một chuỗi S.
Output: In ra "Valid" hoặc "Invalid" cho mỗi
testcase.
Ví dụ:
2
.*\+
.*+
Sample Output:
True
False
Giải thích:
.*+ : lỗi "multiple repeat" - invalid

```python
import re

t = int(input())

for _ in range(t):
    s = input()
    try:
        re.compile(s)
        print(True)
    except:
        print(False)
```
