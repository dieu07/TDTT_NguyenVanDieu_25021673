```python
def reverse(path: str) -> int:
    with open(path, "r", encoding="utf-8") as f:
        s = f.read().strip()
    return int(s[::-1])
```

Viết hàm def canWinNim (path: str) -> bool để xác định xem bạn có thể giành chiến thắng trong trò chơi với số lượng đá đã biết trước không. 

Gợi ý: nếu trong đống đá có 4 hòn đá, thì bạn sẽ không bao giờ thắng được trò chơi: dù bạn lấy 1,2 hay 3 hòn đá, thì bạn của bạn sẽ luôn là người chiến thắng. 

Đầu vào 
Đầu vào của hàm là một xâu biểu diễn đường dẫn đến tệp. Nội dung tệp lưu duy nhất một số nguyên là số lượng đá của trò chơi. 

Đầu ra 
Hàm trả về true nếu bạn có thể thắng, ngược lại trả về false."

```python
def canWinNim(path: str) -> bool:
    with open(path, "r") as f:
        n = int(f.read().strip())     #Nếu số đá ban đầu chia hết cho 4 (tức là n % 4 == 0) → người đi trước thua (bạn sẽ thua nếu đi trước)
    return n % 4 != 0                 #Nếu số đá không chia hết cho 4 (n % 4 == 1, 2 hoặc 3) → người đi trước thắng (bạn có thể thắng)
```

Hai chuỗi là anagram của nhau nếu các chữ cái của chuỗi này có thể được sắp xếp để tạo thành chuỗi kia. Hãy viết hàm def areAnagrams(path1: str, path2: str) trả về true nếu hai chuỗi đầu vào là anagram của nhau ngược lại false. Đầu ra: là True nếu 2 chuỗi kí tự là anagram của nhau, False nếu ngược lại"

```python
def areAnagrams(path1: str, path2: str):
    with open(path1, "r", encoding="utf-8") as f1, open(path2, "r", encoding="utf-8") as f2:
        s1 = f1.read().lower()
        s2 = f2.read().lower()

    a = [c for c in s1 if c.isalpha()]
    b = [c for c in s2 if c.isalpha()]

    return sorted(a) == sorted(b)

if __name__ == "__main__":
    path1 = input().strip()
    path2 = input().strip()
    print(areAnagrams(path1, path2))
```

Mỗi ngôi nhà có một số tiền nhất định, sự ràng buộc duy nhất ngăn cản bạn lấy trộm là hai nhà liền kề có hệ thống an ninh kết nối và nó sẽ tự động liên lạc với cảnh sát nếu hai nhà liền kề đã bị đột nhập vào cùng một đêm. 

Cho danh sách các số nguyên dương biểu diễn cho số tiền của mỗi ngôi nhà, xác định số tiền tối đa bạn có thể trộm tối nay mà không bị báo cảnh sát. 

Viết hàm def getMoney (path: str) -> list[int] để đọc số tiền chứa trong mỗi ngôi nhà từ tệp đầu vào được lưu theo đường dẫn path và trả về mảng chứa các số tiền này 
. Số tiền của mỗi nhà được lưu trong file cách nhau bởi một khoảng trắng và không biết trước số lượng nhà trong khu phố đó. Hàm trả về mảng chứa số tiền của các nhà trong khu phố. 

Viết hàm def rob(hourses: list[int]) -> int để tính và trả về số tiền lớn nhất mà bạn có thể trộm được từ con phố đó mà không bị phát hiện"


```python
def getMoney(path: str):
    with open(path, "r", encoding="utf-8") as f:
        return list(map(int, f.read().strip().split()))

def rob(houses):
    if not houses:
        return 0
    if len(houses) == 1:
        return houses[0]
    dp1, dp2 = houses[0], max(houses[0], houses[1])
    for i in range(2, len(houses)):
        dp1, dp2 = dp2, max(dp2, dp1 + houses[i])
    return dp2

path = input().strip()
houses = getMoney(path)
print(rob(houses))
```

W11A12	"Với một dãy số nguyên cho trước, hãy viết hàm def maximumProduct (path: str) -> int: tìm bộ ba số có tích lớn nhất trong tất cả các bộ số và trả về tích lớn nhất đó.

Hàm nhận đầu vào là đường dẫn đến tệp lưu mảng số nguyên cho trước. Hàm trả về tích lớn nhất có thể có.

Dòng đầu tiên của tệp chứa kích cỡ n, dòng kế tiếp chứa n số nguyên cách nhau bởi một khoảng trắng."

```python
def maximumProduct(path: str):
    with open(path, "r", encoding="utf-8") as f:
        lines = f.read().strip().split()
        n = int(lines[0])
        a = list(map(int, lines[1:]))

    a.sort()
    return max(a[-1]*a[-2]*a[-3], a[0]*a[1]*a[-1])

path = input().strip()
print(maximumProduct(path))
```

W11A13	"Viết hàm def moveZeroes (fileName: str) -> list thực hiện các công việc sau: 

Biết dòng đầu tiên của tệp lưu số nguyên n là số số nguyên dương, dòng tiếp theo chứa n số nguyên dương cần lưu vào mảng. Chuyển tất cả các phần tử của mảng có giá trị bằng 0 về cuối dãy, đồng thời, giữ nguyên vị trí tương đối của các phần tử còn lại trong mảng. Hàm trả lại mảng vừa nhận được 
"

```python
def moveZeroes(fileName: str):
    with open(fileName, "r", encoding="utf-8") as f:
        lines = f.read().strip().split()
        n = int(lines[0])
        a = list(map(int, lines[1:]))

    res = [x for x in a if x != 0]
    res += [0] * (n - len(res))
    return res

fileName = input().strip()
print(moveZeroes(fileName))
```

Lap,Time
1,143.23
2,134.12
Cuối cùng, Frank cần phải tính thời gian trung bình anh hoàn thành vòng đua. Hãy giúp Frank bằng cách viết hàm def averageTime(path: str) , nhận tham số là xâu kí tự lưu đường dẫn tới tệp .csv và trả về thời gian trung bình hoàn thành vòng đua. Lưu ý, tùy tâm trạng Frank sẽ chạy thử xe nhiều vòng hay ít vòng, nên số vòng đua trong các tệp .csv là không cố định. Kết quả lấy đến 2 chữ số phần thập phân."

```python
def averageTime(path: str):
    with open(path, "r", encoding="utf-8") as f:
        lines = f.read().strip().splitlines()[1:]
        times = [float(line.split(",")[1]) for line in lines]
    return round(sum(times) / len(times), 2)

path = input().strip()
print(f"{averageTime(path):.2f}")
```

W11A15	"Thành phố X mở một rạp chiếu phim mới, rất nhiều người muốn đến xem thử và hưởng các khuyến mại trong ngày khai trương.

Để đưa cho khách hàng các lựa chọn, rạp chiếu phim treo các poster về phim cùng với điểm đánh giá IMDB (rating) và mô tả của khách hàng về bộ phim đó.

Hãy viết hàm def findMovies (path: str) -> list[str] giúp khán giả tìm ra phim với điểm đánh giá lớn hơn 8.0 và có mô tả Không phải là ""nhàm chán"" (""boring"").

Hàm nhận đầu vào là đường dẫn đến tệp chứa các poster phim. Hàm trả lại các tên những phim thỏa mãn điều kiện trên. Nếu không tìm thấy phim nào phù hợp, trả về thông báo ""All are boring!"". Kết quả in theo thứ tự xuất hiện trước sau trong file đầu vào

Tệp có cấu trúc như sau:

id, movie, description, rating
1, Thor, great 3D, 8.0
2, The Rings, Fantasy, 7.5
3, Irish, boring, 6.2
4, Ice song, Fantasy, 8.6
5, Home Alone, Interesting, 6.5"

```python
def findMovies(path: str):
    res = []
    with open(path, "r", encoding="utf-8") as f:
        lines = f.read().strip().splitlines()[1:]
        for line in lines:
            parts = [p.strip() for p in line.split(",")]
            movie = parts[1]
            desc = parts[2].lower()
            rating = float(parts[3])
            if rating > 8.0 and desc != "boring":
                res.append(movie)
    if not res:
        return "All are boring!"
    return res

path = input().strip()
result = findMovies(path)
if isinstance(result, list):
    for m in result:
        print(m)
else:
    print(result)
```

W11A17	"Với một dãy gồm n số nguyên với n>1, giá trị của từng phần tử tại vị trí i được tính lại bằng tích của tất cả các phần tử còn lại trong dãy số nguyên trừ chính bản thân nó - ProductofArrayExceptSelf 

Viết hàm def productExceptSelf (path: str) -> list[int]: tính và in ra màn hình dãy số nguyên sau khi được tính lại theo qui tắc trên. 
Hàm nhận đầu vào là đường dẫn path đến tệp chứa dãy số nguyên. Trả lại dãy sau khi biến đổi."	
"VD tệp chứa dãy sau: 
5 
4 6 2 5 2"	

output: 120 80 240 96 240


```python
def productExceptSelf(path: str):
    with open(path, "r", encoding="utf-8") as f:
        lines = f.read().strip().split()
        n = int(lines[0])
        a = list(map(int, lines[1:1+n]))

    total = 1
    zero_count = a.count(0)

    res = []
    if zero_count > 1:
        res = [0] * n
    elif zero_count == 1:
        for x in a:
            if x != 0:
                total *= x
        for x in a:
            if x == 0:
                res.append(total)
            else:
                res.append(0)
    else:
        for x in a:
            total *= x
        for x in a:
            res.append(total // x)

    return res

path = input().strip()
ans = productExceptSelf(path)
print(*ans)
```

W11A18	"Bạn nhận được một tệp văn bản chứa tọa độ điểm xuất phát, tọa độ mục tiêu và vận tốc của một tên lửa (giả sử tên lửa bay thẳng với vận tốc không đổi). Hãy viết một hàm tính thời gian tên lửa cần để bay tới tiêu diệt mục tiêu của mình: def estimatedTime(path: str). Hàm nhận đầu vào là xâu kí tự chứa đường dẫn tới tệp thiết lập tên lửa và trả về số giây tên lửa hoàn thành hành trình. Đơn vị đo khoảng cách là km và đo tốc độ là km/s.

Tệp thiết lập mẫu có cấu trúc như sau:

SRC X: 0 Y: 0
DES X: 10000 Y: 10000
VELOCITY: 6 KM/S

Lấy kết quả tới 2 chữ số phần thập phân"

```python
import math

def estimatedTime(path: str):
    with open(path, "r", encoding="utf-8") as f:
        lines = [line.strip() for line in f if line.strip()]

    x1 = float(lines[0].split()[2])
    y1 = float(lines[0].split()[4])

    x2 = float(lines[1].split()[2])
    y2 = float(lines[1].split()[4])

    v = float(lines[2].split()[1])

    dist = math.sqrt((x2 - x1) ** 2 + (y2 - y1) ** 2)
    return dist / v

path = input().strip()
print(f"{estimatedTime(path):.2f}")
```

W11A19	Viết chương trình nhận vào từ bàn phím một chuỗi kí tự là đường dẫn đến một tệp nhị phân. Tệp này chỉ chứa các số thực (kiểu double). In ra các số thực đó (làm tròn đến 5 chữ số thập phân sau dấu phẩy), mỗi số trên một dòng.

```python
import struct

path = input().strip()

with open(path, "rb") as f:
    while True:
        data = f.read(8)
        if not data:
            break
        x = struct.unpack("d", data)[0]
        print(f"{x:.5f}")
```

W11A20	"Các ảnh màu có thể được chuyển về dạng màu xám (gray scale) với mỗi M×m
điểm ảnh (pixel). Giả sử bạn có một tệp lưu trữ thông tin ảnh trên bằng ma trận số nguyên 2 chiều kích thước M×M.

Nhiệm vụ của bạn là cần thiết kế một hàm để khiến bức ảnh trở nên mượt mà hơn bằng cách tính giá trị của mỗi điểm ảnh bằng giá trị trung bình của tất cả 8
điểm ảnh xung quanh và chính nó (làm tròn xuống). 

Nếu một điểm ảnh có ít hơn 8 điểm ảnh xung quanh, thì hãy sử dụng giá trị trung bình nhiều điểm ảnh xung quanh nhất có thể.

Viết hàm def imageSmoother (path: str) thực hiện công việc trên.

Hàm nhận đầu vào là đường dẫn đến tệp chứa file ảnh. Hàm thực hiện việc làm mịn ảnh và in ra ảnh sau khi làm mịn.

Dòng đầu tiên trong tệp chứa kích thước M của ma trận hai chiều. M dòng tiếp theo mỗi dòng chứa M pixel ứng với thông tin của ảnh.

Chú thích: Giá trị trong ma trận cho trước nằm trong khoảng [0, 255].

Chiều dài và chiều rộng của ma trận cho trước nằm trong khoảng [1, 150]."	
"130 160 150 170 
160 140 180 160 
170 190 120 150 
210 180 160 120"	

"144 140 165 172 
151 155 147 155 
168 177 148 159 
196 201 148 143"

```python
def imageSmoother(path: str):
    with open(path, "r", encoding="utf-8") as f:
        lines = f.read().strip().splitlines()
        M = int(lines[0])
        a = [list(map(int, lines[i+1].split())) for i in range(M)]

    res = [[0]*M for _ in range(M)]
    for i in range(M):
        for j in range(M):
            s = 0
            cnt = 0
            for dx in [-1, 0, 1]:
                for dy in [-1, 0, 1]:
                    ni, nj = i + dx, j + dy
                    if 0 <= ni < M and 0 <= nj < M:
                        s += a[ni][nj]
                        cnt += 1
            res[i][j] = s // cnt

    for row in res:
        print(*row)

path = input().strip()
imageSmoother(path)
```
