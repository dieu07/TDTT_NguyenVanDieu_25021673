Viết hàm merge hai danh sách đã sắp xếp mà không dùng bất cứ
thuật toán sắp xếp nào 

```python
def merge(left, right):
    res = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            res.append(left[i])
            i += 1
        else:
            res.append(right[j])
            j += 1
    res.extend(left[i:])
    res.extend(right[j:])
    return res
a = list(map(int,input().split()))
b = list(map(int,input().split()))
print(merge(a,b))
```

W9A2: Cho ma trận hai chiều nxm có các phần tử được sắp xếp trên
dòng tăng dần và trên cột tăng dần, hãy tìm số lớn thứ k (k <
nxm) với thuật toán tốt nhất có thể 

```python
def kthSmallest(matrix, k):
    n = len(matrix)
    m = len(matrix[0])

    left = matrix[0][0]
    right = matrix[n-1][m-1]

    def count_le(x):
        i = n - 1
        j = 0
        cnt = 0
        while i >= 0 and j < m:
            if matrix[i][j] <= x:
                cnt += i + 1
                j += 1
            else:
                i -= 1
        return cnt

    while left < right:
        mid = (left + right) // 2
        if count_le(mid) < k:
            left = mid + 1
        else:
            right = mid

    return left

n, m = map(int, input().split())
matrix = [list(map(int, input().split())) for _ in range(n)]
k = int(input())

print(kthSmallest(matrix, k))
```

```python
#binary search
def count_ge(matrix, val):
    """
    Đếm số phần tử >= val trong ma trận đã sắp xếp hàng & cột
    Time: O(n + m)
    """
    if not matrix or not matrix[0]:
        return 0
    rows, cols = len(matrix), len(matrix[0])
    count = 0
    r, c = 0, cols - 1  # bắt đầu từ góc phải trên
    
    while r < rows and c >= 0:
        if matrix[r][c] >= val:
            # toàn bộ cột từ 0 đến c đều >= val (vì cột tăng dần xuống dưới)
            count += r + 1
            c -= 1  # đi sang trái
        else:
            r += 1  # đi xuống dưới
    
    return count


def kth_largest(matrix, k):
    if not matrix or not matrix[0]:
        return None
        
    rows, cols = len(matrix), len(matrix[0])
    left = matrix[0][0]          # nhỏ nhất
    right = matrix[-1][-1]       # lớn nhất
    
    # Tìm số lớn thứ k ⇔ tìm số nhỏ nhất mà có đúng (k-1) số lớn hơn nó
    while left < right:
        mid = (left + right + 1) // 2   # bias về phải để tránh lặp vô hạn
        
        # Đếm số phần tử >= mid
        if count_ge(matrix, mid) >= k:
            left = mid
        else:
            right = mid - 1
            
    return left
```

```python
#Tham khảo thêm: Min - heap
from heapq import heappush, heappop

def kth_largest_heap(matrix, k):
    if not matrix or not matrix[0]:
        return None
        
    min_heap = []
    rows, cols = len(matrix), len(matrix[0])
    
    # Bắt đầu từ góc phải dưới (lớn nhất)
    heappush(min_heap, (-matrix[-1][-1], rows-1, cols-1))  # (-giá trị, i, j)
    visited = {(rows-1, cols-1)}
    
    for _ in range(k):
        if not min_heap:
            return None
        val, i, j = heappop(min_heap)
        val = -val  # giá trị thật
        
        # Thêm 2 hướng: lên trên và sang trái (nếu hợp lệ)
        for di, dj in [(-1, 0), (0, -1)]:
            ni, nj = i + di, j + dj
            if 0 <= ni < rows and 0 <= nj < cols and (ni, nj) not in visited:
                visited.add((ni, nj))
                heappush(min_heap, (-matrix[ni][nj], ni, nj))
    
    return val
```

W9A3 Cho một số nguyên dương, hãy thực hiện hoán đổi đúng 2 chữ
số để được một số nguyên nhỏ nhất có thể 

```python
def smallest_after_one_swap(s):
    a = list(s)
    n = len(a)
    for i in range(n):
        min_digit = '9'
        pos = -1

        # Tìm chữ số nhỏ nhất bên phải nhỏ hơn a[i]
        for j in range(i+1, n):
            if a[j] < a[i]:
                # tránh đổi lên đầu thành 0
                if i == 0 and a[j] == '0':
                    continue
                if a[j] <= min_digit:
                    min_digit = a[j]
                    pos = j

        # Nếu tìm được vị trí để đổi
        if pos != -1:
            a[i], a[pos] = a[pos], a[i]
            return ''.join(a)

    # Nếu không tìm được cách làm nhỏ hơn
    # thì đổi 2 số cuối
    a[-1], a[-2] = a[-2], a[-1]
    return ''.join(a)

t = int(input())
for _ in range(t):
    s = input().strip()
    print(smallest_after_one_swap(s))
```

Cho xâu a gồm các chữ cái thường và in hoa và xâu b chỉ gồm
các chữ cái in hoa. Ta có thể biến đổi xâu a theo 2 bước như sau:
- Chuyển một số chữ cái thường trong xâu a thành chữ cái in hoa
(có thể không biến đổi chữ cái nào)
- Xóa hết tất cả các chữ cái thường trong xâu a.
Cho 2 xâu a,b, xác đính xem có cách biến đổi từ xâu a sang xâu
b không. Ghi ra "YES" nếu có và "NO" nếu ngược lại.
Đầu vào:
Dòng đầu tiên chứa 1 số nguyên Q là số lượng câu hỏi. Tương
ứng với mỗi câu hỏi, đưa ra 2 xâu a,b trên 2 dòng khác
nhau.(Q≤10,|a|,|b|≤1000
)
Đầu ra:
Với mỗi câu hỏi, ghi ra kết quả tương ứng trên một dòng.

```python
def biendoi(a, b):
    i = 0  
    j = 0   

    while i < len(a) and j < len(b):
        if a[i].isupper():
            
            if a[i] != b[j]:
                return False
            i += 1
            j += 1
        else:
            # chữ thường
            if a[i].upper() == b[j]:
                # dùng chữ này
                i += 1
                j += 1
            else:
                # bỏ chữ thường này
                i += 1
    if j < len(b):
        return False

    # Kiểm tra phần còn lại của a:
    # nếu còn chữ HOA thì không được xóa
    while i < len(a):
        if a[i].isupper():
            return False
        i += 1

    return True
n = int(input())
for _ in range(n):
    a = input().strip()
    b = input().strip()
    if biendoi(a, b):
        print("YES")
    else:
        print("NO")
```

```python
def can_convert(a, b):
    n = len(a)
    m = len(b)

    # dp[i][j]: dùng a[0..i-1] tạo được b[0..j-1] hay không
    dp = [[False] * (m+1) for _ in range(n+1)]
    dp[0][0] = True

    for i in range(1, n+1):
        for j in range(m+1):

            # Nếu a[i-1] là chữ thường -> có thể xóa
            if a[i-1].islower():
                dp[i][j] = dp[i-1][j]

            # Nếu có thể khớp với b[j-1]
            if j > 0 and a[i-1].upper() == b[j-1]:
                dp[i][j] = dp[i][j] or dp[i-1][j-1]

    return "YES" if dp[n][m] else "NO"

Q = int(input())
for _ in range(Q):
    a = input().strip()
    b = input().strip()
    print(can_convert(a, b))
```

W9A5
Cho hai mảng số nguyên list1 và list2, viết hàm kiểm tra xem list1
có thuộc list2 hay không (có phải là dãy con liên tiếp của list2 hay
không)?

```python
def is_subarray(list1, list2):
    n = len(list1)
    m = len(list2)

    if n > m:
        return False

    for i in range(m - n + 1):
        if list2[i:i+n] == list1:
            return True

    return False
t = int(input())
for _ in range(t):
    list1 = list(map(int, input().split()))
    list2 = list(map(int, input().split()))

    if is_subarray(list1, list2):
        print("YES")
    else:
        print("NO")
```

[Hoán vị tiếp theo]
Cho một dãy số gồm n chữ số là một hoán vị của dãy 1,2,...,n.
Viết chương trình in ra hoán vị ở vị trí tiếp theo (danh sách các
hoán vị được sắp xếp theo thứ tự từ điển) của hoán vị đã cho.
Đầu vào từ bàn phím gồm 2 dòng:
- Dòng đầu tiên chứa số nguyên n
- Dòng tiếp theo chứa n số nguyên là một hoán vị của dãy 1,2,...,n
(n≤105)
Đầu ra: In ra màn hình hoán vị tiếp theo của hoán vị đã cho.

```python
def next_permutation(a):
    n = len(a)

    # Bước 1: tìm i
    i = n - 2
    while i >= 0 and a[i] > a[i+1]:
        i -= 1

    # Nếu là hoán vị cuối cùng
    if i < 0:
        a.reverse()
        return a

    # Bước 2: tìm j
    j = n - 1
    while a[j] < a[i]:
        j -= 1

    # Bước 3: đổi a[i] và a[j]
    a[i], a[j] = a[j], a[i]

    # Bước 4: đảo đoạn sau i
    a[i+1:] = reversed(a[i+1:])

    return a

n = int(input())
a = list(map(int, input().split()))

res = next_permutation(a)
print(*res)
```

```python
def canPlaceFlowers(flowerbed, k):
    n = len(flowerbed)

    for i in range(n):
        if flowerbed[i] == 0:
            left = (i == 0 or flowerbed[i-1] == 0)
            right = (i == n-1 or flowerbed[i+1] == 0)

            if left and right:
                flowerbed[i] = 1   # trồng hoa
                k -= 1
                if k == 0:
                    return True

    return k <= 0
t = int(input())
for _ in range(t):
    flowerbed = list(map(int, input().split()))
    k = int(input())

    if canPlaceFlowers(flowerbed, k):
        print("True")
    else:
        print("False")
```

[Phát hiện đường đi lặp lại của Rô-bốt]
Ban đầu, một con rô-bốt ở vị trí (0, 0) trên một lưới 2D.
Cho một chuỗi `moves` chứa các thao tác di chuyển của rô-bốt.
Các bước di chuyển được biểu diễn bởi các ký tự:
- 'R' (Phải)
- 'L' (Trái)
- 'U' (Lên)
- 'D' (Xuống)
Viết hàm `def is_path_crossing(moves)`:
Hàm nhận đầu vào là một chuỗi `moves`.
Hàm trả về `True` nếu trong quá trình di chuyển, rô-bốt đi vào một
ô mà nó đã từng đi qua trước đó.
Hàm trả về `False` nếu rô-bốt không bao giờ đi lặp lại (luôn đi vào
ô mới).
Lưu ý: Vị trí ban đầu (0, 0) được tính là vị trí đầu tiên đã đi qua.

```python
def is_path_crossing(moves):
    x, y = 0, 0
    visited = set()
    visited.add((0, 0))

    for move in moves:
        if move == 'R':
            x += 1
        elif move == 'L':
            x -= 1
        elif move == 'U':
            y += 1
        elif move == 'D':
            y -= 1

        # Nếu đã từng tới ô này
        if (x, y) in visited:
            return True

        visited.add((x, y))

    return False
t = int(input())
for _ in range(t):
    moves = input().strip()
    if is_path_crossing(moves):
        print("True")
    else:
        print("False")
```

Viết hàm `def Median(nums):`
Hàm nhận đầu vào là một danh sách (list) `nums` được đảm bảo
có đúng 5 số nguyên.
Yêu cầu:
Tìm và trả về số trung vị (số lớn thứ ba) của 5 số này.
Ràng buộc:
1. Chỉ được sử dụng tối đa 6 phép so sánh (ví dụ: <, >, <=, >=).
2. KHÔNG ĐƯỢC DÙNG bất kỳ hàm sắp xếp nào (như
`sorted()`, `list.sort()`).
Chú ý: Bạn chỉ cần viết hàm `Median` theo yêu cầu.
Phần đọc input và gọi hàm (hàm main) đã có sẵn.
1 2 3 4 5
5 4 3 2 1
1 5 3 4 2
1 10 3 8 5
1 4 3 8 5
4
3
3
10
5 

```python
def Median(nums):
    a, b, c, d, e = nums

    # So sánh 1
    if a > b:
        a, b = b, a

    # So sánh 2
    if c > d:
        c, d = d, c

    # So sánh 3
    if a > c:
        a, c = c, a
        b, d = d, b

    # Lúc này: a <= c và b >= d

    # So sánh 4
    if b < c:
        mid1 = b
    else:
        mid1 = c

    # So sánh 5
    if mid1 < e:
        # So sánh 6
        if e < max(b, d):
            return e
        else:
            return mid1
    else:
        # So sánh 6
        if max(a, c) < e:
            return e
        else:
            return mid1
```

[Tìm bán kính lò sưởi nhỏ nhất]
Mùa đông đang đến! Bạn được cung cấp vị trí của các ngôi nhà
và các lò sưởi trên một đường thẳng.
Nhiệm vụ của bạn là tìm bán kính r nhỏ nhất cho các lò sưởi sao
cho mọi ngôi nhà đều nằm trong phạm vi sưởi ấm của ít nhất một
lò sưởi.
Tất cả các lò sưởi đều có cùng một bán kính r.
Một ngôi nhà tại vị trí h được sưởi ấm bởi một lò sưởi tại vị trí t
nếu |h - t| <= r.
Viết hàm `def findRadius(houses, heaters):`
Hàm nhận đầu vào là hai danh sách (list) các số nguyên:
- `houses`: chứa vị trí của các ngôi nhà.
- `heaters`: chứa vị trí của các lò sưởi.
Hàm trả về một số nguyên là bán kính r nhỏ nhất cần thiết.
Lưu ý: 
- Số lượng nhà và lò sưởi là một số không âm và không vượt quá
25000.
- Vị trí của các ngôi nhà và lò sưởi là một số không âm và không
vượt quá 10^9. 

```python
def findRadius(houses, heaters):
    houses.sort()
    heaters.sort()

    j = 0
    res = 0

    for h in houses:
        # dịch con trỏ sao cho heaters[j] là lò gần nhất bên phải hoặc đúng
        while j + 1 < len(heaters) and abs(heaters[j + 1] - h) <= abs(heaters[j] - h):
            j += 1

        # khoảng cách từ nhà h tới lò gần nhất
        dist = abs(heaters[j] - h)

        # cập nhật bán kính lớn nhất cần thiết
        res = max(res, dist)

    return res
n = int(input())
houses = list(map(int, input().split()))
m = int(input())
heaters = list(map(int, input().split()))

print(findRadius(houses, heaters))
```
