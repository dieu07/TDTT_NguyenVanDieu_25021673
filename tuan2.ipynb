```python
#Bài 5
A = input("Nhập vào chữ cái hoa:")
if A.isupper():
    if A == "A":
        print(f"Chữ cái thường liền trước {A} là: z")
    else:
        print(f"Chữ cái thường liền trước {A} là: {chr(ord(A)+32-1)}")
elif A.islower():
    print(f"{A} không phải là chữ hoa")
else:
    print(f"{A} không phải là chữ alphabet")
```

```python
#Bai 6
import math
a = float(input("Nhập cạnh a: "))
b = float(input("Nhập cạnh b: "))
c = float(input("Nhập cạnh c: "))
if a + b > c and a + c > b and b + c > a:
    p = (a + b + c) / 2
    S = math.sqrt(p * (p - a) * (p - b) * (p - c)) #Ngoài ra có thể dùng lũy thừa 1/2 thay căn **0.5
    print(f"Diện tích tam giác là: {S:.1f}")
else:
    print("Khong phai 3 canh cua tam giac")
```

```python
#Bai 8
ten = input("Nhập họ và tên:")
cu = float(input("Nhập chỉ số tháng trước:"))
moi = float(input("Nhập chỉ số tháng này: "))
so_dien = moi - cu
if so_dien < 0:
    print("Nhập sai! Nhập lại!")
else:
    if so_dien <= 50:
        tien = so_dien * 1984
    elif so_dien <= 100:
        tien = 50 * 1984 + (so_dien - 50) * 2050
    elif so_dien <= 200:
        tien = 50 * 1984 + 50 * 2050 + (so_dien - 100)*2380
    elif so_dien <= 300:
        tien = 50 * 1984 + 50 * 2050 + 100*2380 + (so_dien - 200)*2998
    elif so_dien <= 400:
        tien = 50*1984 + 50*2050 + 100*2380 + 100*2998 +(so_dien - 300)*3350
    else:
        tien = 50*1984 + 50*2050 + 100*2380 + 100*2998 + 100*3350 + (so_dien - 400)*3460
tong = tien * ( 1 + 0.08 )
print(f"Ho va ten: {ten}\nTien phai tra la: {tong:.0f} ")
```
