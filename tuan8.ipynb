Đầu vào

Đầu ra
In ra màn hình diện tích bề mặt và thể tích của
hình trụ (sử dụng lại những hàm vừa viết, π=3.
14).
(Kết quả làm tròn đến 2 chữ số sau dấu phẩy).

```python
class Cylinder:
    def __init__(self, r, h):
        self.r = r
        self.h = h
    def dien_tich(self):
        return 2 * self.r * 3.14 * self.h + 2 * (self.r**2) * 3.14
    def the_tich(self):
        return 3.14 * (self.r**2) * self.h
    def __str__(self):
        return f"{self.dien_tich():.2f} {self.the_tich():.2f}"
r, h = map(int,input().split())
p = Cylinder(r, h)
print(p)
    
```

Với các phương thức sau:
1. Khởi tạo với danh sách rỗng
2. Thêm vào giỏ hàng một sản phẩm mới
3. Xóa khỏi giỏ một sản phẩm nào đó (tìm theo tên

nhập "yes" cho câu hỏi "Có muốn thoát khỏi
chương trình không?"


```python
class ShoppingCart:
    def __init__(self):
        self.items = []
    def add_item(self, name, price):
        self.items.append({"name":name, "price" : price})
    def remove_item(self, name):
        for item in self.items:
            if item["name"] == name:
                self.items.remove(item)
                return True
        return False
    def is_empty(self):
        return len(self.items) == 0
    def total_price(self):
        return sum(item["price"] for item in self.items)
    def display_items(self):
        if self.is_empty():
            print("Giỏ hàng rỗng")
        else:
            for item in self.items:
                print(f"Tên: {item["name"]}, Giá: {item["price"]}")
    def clear_cart(self):
        self.items = []
cart = ShoppingCart()
while True:
    print("----Shopping Cart----")
    print("""
2. Thêm vào giỏ hàng một sản phẩm mới
3. Xóa khỏi giỏ một sản phẩm nào đó (tìm theo tên
để xóa)
4. Kiểm tra xem giỏ hàng có rỗng không
5. Tính tổng tiền giỏ hàng
6. Hiển thị toàn bộ sản phẩm trong giỏ
7. Xóa toàn bộ giỏ hàng""")
    choice = input("Chọn 2-7:")
    if choice == "2":
        name = input("Nhập tên sản phẩm:")
        price = float(input("Nhập giá:"))
        cart.add_item(name, price)
        print("Thêm sản phẩm thành công!")
    elif choice == "3":
        name = input("Nhập tên sp cần xóa:")
        if cart.remove_item(name):
            print("Sản phẩm đã được xóa!")
        else:
            print("Không tìm thấy sản phẩm!")
    elif choice == "4":
        if cart.is_empty():
            print("Giỏ hàng rỗng")
        else:
            print("Giỏ hàng không rỗng")
    elif choice == "5":
        print(f"Tổng tiền là: {cart.total_price()}")
    elif choice == "6":
        cart.display_items()
    elif choice == "7":
        cart.clear_cart()
        print("Giỏ hàng đã xóa hết")
    else:
        print("không hợp lệ")
    exit_choice = input("Bạn có muốn thoát? Nhập yes/ no:").strip().lower()
    if exit_choice == "yes":
        break
```

Xây dựng lớp Rectangle(w, h) với các phương
thức:
area() trả về diện tích.
perimeter() trả về chu vi.
scale(k) phóng to hình chữ nhật theo hệ số k
(nhân cả w và h), không trả về gì.
Input
w h k (số thực dương)
Output
area perimeter sau khi phóng to (làm tròn 2 chữ
số thập phân).

```python
class Rectangle:
    def __init__(self, w, h):
        self.w = w
        self.h = h
    
    def area(self):
        return self.w * self.h
    
    def perimeter(self):
        return 2 * (self.w + self.h)
    
    def scale(self, k):
        self.w *= k
        self.h *= k


try:
    w, h, k = map(float, input().split())
    if w <= 0 or h <= 0 or k <= 0:
        print("Phải nhập số dương")
    else:
        rect = Rectangle(w, h)
        rect.scale(k)
        print(f"{rect.area():.2f}")
        print(f"{rect.perimeter():.2f}")

except ValueError:
    print("Dữ liệu không đúng định dạng")
```

W8A7
Yêu cầu
Lớp Fraction(num, den) (tử, mẫu nguyên, mẫu ≠
0) với:
Rút gọn phân số trong __init__ (quy ước mẫu >
0).
Các phương thức add/sub/mul/div trả về phân số
đã rút gọn.
__str__() in dạng a/b.
Input
a b op c d (tương ứng a/b (op) c/d, op thuộc {+,-,*,
/})
Output
Phân số kết quả đã rút gọn

```python
import math

class Fraction:
    def __init__(self, num, den):
        if den == 0:
            raise ValueError("Mẫu số không được bằng 0")
        
        # Rút gọn ngay trong __init__
        g = math.gcd(num, den)          # gcd xử lý cả số âm
        self.num = num // g
        self.den = den // g
        
        # Quy ước mẫu > 0
        if self.den < 0:
            self.num = -self.num
            self.den = -self.den
    
    def add(self, other):
        new_num = self.num * other.den + other.num * self.den
        new_den = self.den * other.den
        return Fraction(new_num, new_den)
    
    def sub(self, other):
        new_num = self.num * other.den - other.num * self.den
        new_den = self.den * other.den
        return Fraction(new_num, new_den)
    
    def mul(self, other):
        new_num = self.num * other.num
        new_den = self.den * other.den
        return Fraction(new_num, new_den)
    
    def div(self, other):
        if other.num == 0:
            raise ValueError("Không thể chia cho 0")
        new_num = self.num * other.den
        new_den = self.den * other.num
        return Fraction(new_num, new_den)
    
    def __str__(self):
        if self.den == 1:
            return str(self.num)
        return f"{self.num}/{self.den}"

try:
    # Nhập 1 dòng: a b op c d
    line = input().strip()
    parts = line.split()
    
    if len(parts) != 5:
        raise ValueError("Input không đúng định dạng")
    
    a = int(parts[0])
    b = int(parts[1])
    op = parts[2]
    c = int(parts[3])
    d = int(parts[4])
    
    f1 = Fraction(a, b)
    f2 = Fraction(c, d)
    
    if op == '+':
        result = f1.add(f2)
    elif op == '-':
        result = f1.sub(f2)
    elif op == '*':
        result = f1.mul(f2)
    elif op == '/':
        result = f1.div(f2)
    else:
        raise ValueError("Toán tử không hợp lệ")
    
    print(result)
    
except ValueError as e:
    print("Dữ liệu nhập không hợp lệ")
except Exception as e:
    print("Lỗi:", str(e))
```

Yêu cầu
Lớp BankAccount(owner, balance) với:
deposit(x), withdraw(x); nếu rút quá số dư thì bỏ
qua giao dịch.
__str__() in owner: balance.
Input
Dòng 1: owner initial_balance
Dòng 2: n (số lệnh)
Tiếp theo n dòng, mỗi dòng: DEPOSIT x hoặc
WITHDRAW x.
Output
Số dư cuối cùng (in số thực, 2 chữ số).

```python
class BankAccount:
    def __init__(self,owner,balance):
        self.owner = owner
        self.balance = balance
    def deposit(self, x):
        self.balance += x
    def withdraw(self, x):
        if self.balance - x < 0:
            raise ValueError("Khong the thuc hien giao dich vi khong du so du")
        else:
            self.balance -= x
    def __str__(self):
        return f"{self.balance:.2f}"
nhap = input().strip().split()
owner = nhap[0]
balance = float(nhap[1])
bank = BankAccount(owner, balance)
n = int(input("Số giao dịch:"))
try:
    for i in range(n):
        x = input("Nhập lệnh DEPOSIT x hoặc WITHDRAW x:").strip().split()
        if x[0].lower() == "deposit":
            bank.deposit(float(x[1]))
        elif x[0].lower() == "withdraw":
            bank.withdraw(float(x[1]))
        else:
            print("Loi cu phap")
    print(bank)
except ValueError as e:
    print("Lỗi:",str(e))
except Exception as e:
    print("Lỗi:", str(e))
```

W8A9 Yêu cầu
Lớp Student(name) với:
add_score(subject, score)
avg() trả về điểm trung bình.
rank() trả về "Excellent" nếu avg ≥ 8, "Good" nếu
6.5–<8, "Average" nếu 5–<6.5, ngược lại "Poor".
Input
Dòng 1: name
Dòng 2: n
n dòng tiếp: subject score (score là số thực 0–
10)
Output
name avg rank (avg 2 chữ số).

```python
class Student:
    def __init__(self, name):
        self.name = name
        self.scores = {}
    
    def add_score(self, subject, score):
        if 0 <= score <= 10:
            self.scores[subject] = score
    def avg(self):
        if not self.scores:
            return 0.0
        total = sum(self.scores.values())
        return total / len(self.scores)
    
    def rank(self):
        average = self.avg()
        if average >= 8:
            return "Excellent"
        elif average >= 6.5:
            return "Good"
        elif average >= 5:
            return "Average"
        else:
            return "Poor"
try:
    name = input().strip()
    n = int(input().strip())
    student = Student(name)
    for _ in range(n):
        line = input().strip().split()
        if len(line) < 2:
            continue  # bỏ qua dòng sai
        
        # subject có thể nhiều từ, score là số cuối
        subject = ' '.join(line[:-1])
        try:
            score = float(line[-1])
            student.add_score(subject, score)
        except ValueError:
            continue  # bỏ qua nếu score không phải số
    average = student.avg()
    ranking = student.rank()
    
    print(f"{name} {average:.2f} {ranking}")
except Exception:
    print("Dữ liệu nhập không hợp lệ")
```

Mỗi lệnh COUNT/COUNTYEAR in một dòng kết
quả.


```python
class Book:
    def __init__(self, title, author, year):
        self.title = title
        self.author = author
        self.year = year


class Library:
    def __init__(self):
        self.books = []  # danh sách các đối tượng Book
    
    def add(self, book):
        self.books.append(book)
    
    def count_by_author(self, name):
        count = 0
        for book in self.books:
            if book.author == name:
                count += 1
        return count
    
    def find_by_year(self, y):
        count = 0
        for book in self.books:
            if book.year == y:
                count += 1
        return count


# Phần xử lý input và output
try:
    n = int(input().strip())
    
    library = Library()
    
    for _ in range(n):
        line = input().strip()
        
        # Tách lệnh và dữ liệu
        if line.startswith("ADD "):
            # ADD title;author;year
            data = line[4:].split(';')  # bỏ "ADD " rồi tách theo ;
            if len(data) != 3:
                continue  # bỏ qua dòng sai định dạng
            
            title = data[0].strip()
            author = data[1].strip()
            try:
                year = int(data[2].strip())
                book = Book(title, author, year)
                library.add(book)
            except ValueError:
                continue  # year không phải số → bỏ qua
        
        elif line.startswith("COUNT "):
            # COUNT author_name
            author_name = line[6:].strip()
            result = library.count_by_author(author_name)
            print(result)
        
        elif line.startswith("COUNTYEAR "):
            # COUNTYEAR year
            try:
                year = int(line[10:].strip())
                result = library.find_by_year(year)
                print(result)
            except ValueError:
                print(0)  # năm không hợp lệ → coi như 0
        
        else:
            # Lệnh không hợp lệ → bỏ qua
            pass

except Exception:
    # Nếu có lỗi lớn (ví dụ n không phải số)
    pass  # theo đề không yêu cầu in lỗi, chỉ xử lý bình thường
```

W8A11
Bạn được cho bốn điểm A, B, C và D trong hệ tọa
độ Đề-các 3 chiều.
Nhiệm vụ của bạn là tính và in ra góc giữa hai mặt
phẳng tạo bởi các điểm:
mặt phẳng (A, B, C)
mặt phẳng (B, C, D)
Góc cần in ra được ký hiệu là PHI, và phải dùng
độ (degrees), không dùng radian.
Dạng Input:
Một dòng input chứa 3 số thực, ngăn cách bằng
dấu cách, biểu diễn:
tọa độ X tọa độ Y tọa độ Z của một điểm.
Dạng Output:
In ra góc giữa hai mặt phẳng, chính xác tới hai
chữ số thập phân.

```python
import math

class Point:
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z
    
    def __sub__(self, other):
        return Vector(
            self.x - other.x,
            self.y - other.y,
            self.z - other.z
        )

class Vector:
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z
    
    def dot(self, other):
        return (self.x * other.x +
                self.y * other.y +
                self.z * other.z)
    
    def cross(self, other):
        return Vector(
            self.y * other.z - self.z * other.y,
            self.z * other.x - self.x * other.z,
            self.x * other.y - self.y * other.x
        )
    
    def magnitude(self):
        return math.sqrt(self.x**2 + self.y**2 + self.z**2)


# Đọc 4 điểm (mỗi điểm 1 dòng)
points = []
for _ in range(4):
    line = input().strip()
    if not line:
        continue
    x, y, z = map(float, line.split())
    points.append(Point(x, y, z))

A, B, C, D = points

# Vector AB, AC cho mặt phẳng ABC
AB = B - A
AC = C - A

# Vector BC, BD cho mặt phẳng BCD
BC = C - B
BD = D - B

# Vector pháp tuyến n1 của mặt phẳng ABC
n1 = AB.cross(AC)

# Vector pháp tuyến n2 của mặt phẳng BCD
n2 = BC.cross(BD)

# Tính cos phi = |n1 · n2| / (|n1| * |n2|)
dot_prod = abs(n1.dot(n2))
mag_n1 = n1.magnitude()
mag_n2 = n2.magnitude()

if mag_n1 == 0 or mag_n2 == 0:
    # Một trong hai mặt phẳng suy biến (3 điểm thẳng hàng)
    phi_deg = 0.00
else:
    cos_phi = dot_prod / (mag_n1 * mag_n2)
    # Giới hạn cos_phi trong [-1, 1] để tránh lỗi số học
    cos_phi = max(min(cos_phi, 1.0), -1.0)
    phi_rad = math.acos(cos_phi)
    phi_deg = math.degrees(phi_rad)

# In kết quả với 2 chữ số thập phân
print(f"{phi_deg:.2f}")
```

[Complex]
Khai báo kiểu dữ liệu biểu diễn số phức \
(ComplexNumber\) bao gồm \(2\) biến \(realPart\)
và \(imaginaryPart\) kiểu \(double\) biểu diễn phần
thực và phần ảo của một số phức.
Viết hàm khởi tạo ComplexNumber(double
initRealPart, double initImaginaryPart) để khởi tạo
các giá trị phần thực và phần ảo của một số phức
và hàm ComplexNumber multiply(complexNumber
other) trả về tích của số phức hiện tại và một số
phức khác.
Đầu vào
Đầu vào từ bàn phím gồm hai dòng, mỗi dòng
gồm hai số thực cách nhau bởi một dấu cách biểu
diễn phần thực và phần ảo của một số phức. Các
số trong đầu vào có giá trị tuyệt đối không vượt
quá \(100\).
Đầu ra
In ra màn hình hai số biểu diễn phần thực và phần
ảo của tích hai số phức.

```python
class ComplexNumber:
    def __init__(self, initRealPart, initImaginaryPart):
        self.realPart = initRealPart
        self.imaginaryPart = initImaginaryPart
    
    def multiply(self, other):
        real = (self.realPart * other.realPart -
                self.imaginaryPart * other.imaginaryPart)
        imag = (self.realPart * other.imaginaryPart +
                self.imaginaryPart * other.realPart)
        return ComplexNumber(real, imag)


try:
    a, b = map(float, input().split())
    c, d = map(float, input().split())
    
    z1 = ComplexNumber(a, b)
    z2 = ComplexNumber(c, d)
    
    result = z1.multiply(z2)
    
    # In với 2 chữ số thập phân
    print(f"{result.realPart:.2f} {result.imaginaryPart:.2f}")

except:
    pass
```

[Triangle]
Khai báo cấu trúc dữ liệu tam giác \(Triangle\)
chứa \(3\) biến \(egde1,\,edge2,\,edge3\) kiểu \
(int\) là độ dài \(3\) cạnh của tam giác đó.
Viết hàm khởi tạo Triangle(int initEdge1,int
initEdge2, int initEdge3) để khởi tạo độ dài các
cạnh của tam giác và hàm double getArea() trả về
diện tích của tam giác đó (nếu tam giác không
hợp lệ, trả về \(-1\)).
Hoàn thiện chương trình dưới đây nhập vào \(3\)
cạnh của một tam giác và in ra diện tích của tam
giác đó.
Đầu vào
Một dòng duy nhất từ bàn phím chứa \(3\) số
nguyên là độ dài \(3\) cạnh của tam giác
Đầu ra
In ra màn hình diện tích tam giác (làm tròn đến\
(2\) chữ số thập phân sau dấu phẩy). In ra \
(invalid\) nếu \(3\) số nhập vào không phải là \(3\)
cạnh của một tam giác


```python
import math

class Triangle:
    def __init__(self, a, b, c):
        self.a, self.b, self.c = a, b, c
    
    def area(self):
        if min(self.a, self.b, self.c) <= 0:
            return -1
        if self.a + self.b <= self.c or self.a + self.c <= self.b or self.b + self.c <= self.a:
            return -1
        p = (self.a + self.b + self.c) / 2
        return math.sqrt(p * (p - self.a) * (p - self.b) * (p - self.c))

t = Triangle(*map(int, input().split()))
res = t.area()
print("invalid" if res < 0 else f"{res:.2f}")
```
