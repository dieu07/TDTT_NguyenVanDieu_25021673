```python
#b9
n = int(input())

def chu_so_khac_nhau(x):
    s = str(x)
    return len(s) == len(set(s))

found = False
for i in range(1, int(n**0.5) + 1):
    sq = i * i
    if sq > n:
        break
    if chu_so_khac_nhau(sq):
        if found:
            print(" ", end="")
        print(sq, end="")
        found = True

if not found:
    print("no number")
else:
    print()      # xuống dòng cuối cùng
```

```python
#b10
n = int(input())

memo = {1: 1}  # độ dài dãy Collatz của 1 là 1

def collatz_len(x):
    if x in memo:
        return memo[x]
    if x % 2 == 0:
        memo[x] = 1 + collatz_len(x // 2)
    else:
        memo[x] = 1 + collatz_len(3 * x + 1)
    return memo[x]

best_x = 1
best_len = 1

for x in range(1, n + 1):
    L = collatz_len(x)
    if L > best_len or (L == best_len and x < best_x):
        best_x = x
        best_len = L

print(best_x, best_len)
```

```python
#b13
def sum_divisors(n):
    s = 1  # 1 luôn là ước (trừ n=1 nhưng không ảnh hưởng)
    # duyệt đến sqrt(n)
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            s += i
            if i != n // i:
                s += n // i
    return s
a, b = map(int, input().split())

if a > 1 and b > 1 and sum_divisors(a) == b and sum_divisors(b) == a:
    print("true")
else:
    print("false")
```

```python
#b22:Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ
n = int(input("Nhập số nguyên dương n: "))

chan = 0
le = 0

while n > 0:
    digit = n % 10      # lấy chữ số cuối
    if digit % 2 == 0:
        chan += 1
    else:
        le += 1
    n //= 10            # bỏ chữ số cuối

print("Số chữ số chẵn:", chan)
print("Số chữ số lẻ:", le)
```

```python
#b25: Nhập vào một dãy số nguyên, ngưng nhập khi người dùng nhập -1. Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhậpprint("Nhập các số nguyên (nhập -1 để kết thúc):")

first_input = True
while True:
    num = int(input())
    if num == -1:
        break
    if first_input:
        max_num = num
        min_num = num
        first_input = False
    else:
        if num > max_num:
            max_num = num
        if num < min_num:
            min_num = num

if first_input:
    print("Bạn chưa nhập số nào!")
else:
    print("Số lớn nhất:", max_num)
    print("Số nhỏ nhất:", min_num)
```

```python
#b26: Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ bằng tổng hai số trước đó
#Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt quá AA = int(input("Nhập số nguyên dương A: "))

A = int(input("Nhập số nguyên dương A: "))

# Khởi tạo hai số đầu tiên
a, b = 1, 1

# Nếu A < 1, không có số phù hợp
if A < 1:
    print("Không có số Fibonacci nào ≤ A")
else:
    while b <= A:
        a, b = b, a + b

    # a là số Fibonacci lớn nhất ≤ A
    print("Số Fibonacci lớn nhất không vượt quá A là:", a)
```

```python
#b27: Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu
#từ (quy định là chuỗi không có ký tự đặc biệt, không số,
#không có dấu câu, chỉ có ký tự chữ và khoảng trắng)
#Dùng cách: for trên index; KHÔNG dùng hàm split(), len()
s = input("Nhập chuỗi: ")

count = 0
in_word = False

for i in range(len(s)):  # vòng lặp for trên index
    if s[i] != ' ' and not in_word:
        # Bắt đầu một từ mới
        in_word = True
        count += 1
    elif s[i] == ' ':
        # Kết thúc từ hiện tại
        in_word = False

print("Số từ trong chuỗi là:", count)
```

```python
#b28: Nhập vào một chuỗi, hãy in từ đầu tiên trong chuỗi
#Dùng cách: for trên index; KHÔNG dùng hàm split()
s = input("Nhập chuỗi: ")

first_word = ""
found_word = False

for i in range(len(s)):
    if s[i] != ' ':
        first_word += s[i]
        found_word = True
    elif s[i] == ' ' and found_word:
        # đã kết thúc từ đầu tiên
        break

if first_word:
    print("Từ đầu tiên trong chuỗi là:", first_word)
else:
    print("Chuỗi không có từ nào")
```

```python
#b30: Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu
#ký tự in hoa, bao nhiêu ký tự in thường, bao nhiêu ký tự số
s = input("Nhập chuỗi: ")

chu_hoa = 0
chu_thuong = 0
chu_so = 0

for ch in s:
    if 'A' <= ch <= 'Z':
        chu_hoa += 1
    elif 'a' <= ch <= 'z':
        chu_thuong += 1
    elif '0' <= ch <= '9':
        chu_so += 1

print("Số ký tự in hoa:", chu_hoa)
print("Số ký tự in thường:", chu_thuong)
print("Số ký tự số:", chu_so)
```

```python
#b32: Nhập vào một chuỗi, kiểm tra chuỗi đó có phải là một chuỗi
#mật khẩu mạnh hay không (một chuỗi mật khẩu mạnh cần có
#ít nhất 1 ký tự đặc biệt, 1 ký tự in hoa, 1 con số, 1 chữ
#thường và độ dài phải lớn hơn 6)
s = input("Nhập chuỗi mật khẩu: ")

has_upper = False
has_lower = False
has_digit = False
has_special = False

if len(s) <= 6:
    print("Mật khẩu yếu: độ dài phải > 6")
else:
    for ch in s:
        if 'A' <= ch <= 'Z':
            has_upper = True
        elif 'a' <= ch <= 'z':
            has_lower = True
        elif '0' <= ch <= '9':
            has_digit = True
        else:
            has_special = True

    if has_upper and has_lower and has_digit and has_special:
        print("Mật khẩu mạnh")
    else:
        print("Mật khẩu yếu")
```

```python
#b33:Nhập vào một số nguyên, hãy chuyển số sang chuỗi, rồi đặt
#dấu chấm phân tách mỗi 3 chữ số (phân cách phần ngàn) rồi
#in ra màn hình
#VD:
#Nhập số: 375469485
#Đổi sang chuỗi rồi in ra: 375.469.485
num = int(input("Nhập số: "))

num_str = str(num)
result = ""

count = 0

# Duyệt từ cuối chuỗi về đầu
for i in range(len(num_str)-1, -1, -1):
    result = num_str[i] + result
    count += 1
    if count % 3 == 0 and i != 0:
        result = "." + result

print("Đổi sang chuỗi rồi in ra:", result)
```

```python
#b33 slicing
num = int(input("Nhập số: "))
num_str = str(num)

# Khởi tạo kết quả
result = ""

# Tính vị trí phần dư để lấy nhóm đầu tiên (nếu số chữ số không chia hết cho 3)
remainder = len(num_str) % 3
if remainder != 0:
    result = num_str[:remainder]

# Duyệt từng nhóm 3 chữ số còn lại
for i in range(remainder, len(num_str), 3):
    if result != "":
        result += "."
    result += num_str[i:i+3]

print("Đổi sang chuỗi rồi in ra:", result)
```

```python
#b34:Nhập vào chuỗi a và chuỗi b
#Hãy xóa chuỗi b trong chuỗi a rồi in lại chuỗi a ra màn hình (không dùng hàm replace)
#Ví dụ:
#Chuỗi a: "Xin chào mọi người!"
#Chuỗi b: "Xin chào"
#Sau khi xóa, chuỗi a: " mọi người!"
a = input("Nhập chuỗi a: ")
b = input("Nhập chuỗi b: ")

result = ""
i = 0
while i < len(a):
    # Kiểm tra xem từ vị trí i, có trùng chuỗi b không
    if a[i:i+len(b)] == b:
        i += len(b)  # bỏ qua chuỗi b
    else:
        result += a[i]
        i += 1

print("Sau khi xóa, chuỗi a:", result)
```
